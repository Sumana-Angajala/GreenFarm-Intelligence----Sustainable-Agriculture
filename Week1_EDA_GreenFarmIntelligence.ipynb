```python
# ðŸ“Œ Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ðŸ“Œ Step 2: Load Dataset
df = pd.read_csv(r"C:\Users\suman\Downloads\archive (4)\Crop_recommendation.csv")  # make sure the CSV is in the same folder
print("âœ… Dataset Loaded Successfully!")
print("Shape:", df.shape)

# ðŸ“Œ Step 3: Preview Dataset
df.head()
```

```python
# ðŸ“Œ Step 4: Dataset Info
df.info()
df.describe()
```

```python
# ðŸ“Œ Step 5: Check Missing Values
print("Missing Values per Column:")
print(df.isnull().sum())
```

```python
# ðŸ“Œ Step 6: Feature Distributions
features = ['N','P','K','temperature','humidity','ph','rainfall']

for feat in features:
    plt.figure(figsize=(6,3))
    sns.histplot(df[feat], kde=True, bins=20, color="green")
    plt.title(f"Distribution of {feat}")
    plt.show()

```

```python
# ðŸ“Œ Step 7: Correlation Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df[features].corr(), annot=True, cmap="YlGnBu")
plt.title("Correlation Heatmap of Features")
plt.show()
```

```python
# ðŸ“Œ Step 8: Crop Distribution
plt.figure(figsize=(10,5))
df['label'].value_counts().plot(kind="bar", color="seagreen")
plt.title("Crop Distribution")
plt.xlabel("Crop")
plt.ylabel("Count")
plt.show()
```
